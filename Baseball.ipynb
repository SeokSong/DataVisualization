```python
def read_data(file):
    lst_of_lst = [] #empty list to append too
    header = file.readline().strip().split(",") # Get a list for the header incase it is needed later
    for line in file: #for loop through file
        lst = line.strip().split(",") #make a list for each row 
        for i in range(0, len(lst)): #for loop to make sure each value is the correct value
            value = lst[i].isnumeric() #check if the value is all numbers 
            if value == True:
                lst[i] = int(lst[i]) #if it is True make it a int, removes floats
            elif lst[i] == '':
                lst[i] = 0 #if the value is empty make it 0
        # dont have to do string because it if False it will already be a string
        lst_of_lst.append(lst) #append list of list to lst_of_lst
    return lst_of_lst
```

```python
def get_matching_rows(lst_of_lst, index_column, target_value):
    matching_rows_lst = [] #empty list too append too
    for i in lst_of_lst: #for loop through the list of list
        if i[index_column] == target_value: #check if the index column at that row == target value
            matching_rows_lst.append(i) #append the row if the index column for that row == target value
    return(matching_rows_lst)
#gets all the data from 1 year into a seperate list
```

```python
def get_column_values(lst_of_lst, index_column):
    column_values_lst = [] #new list to append too
    for i in lst_of_lst: #for loop
        column_values_lst.append(i[index_column]) #this will append the values inside the index column requested
    return column_values_lst
#gets all the values from a requested column
```

```python
def get_unique_values(lst_of_values):
    unique_values_lst = [] #new list to append too
    for i in lst_of_values: #for loop
        if i not in unique_values_lst: #this will only append if the value from the given list is not already in the new list
            unique_values_lst.append(i)
    return(unique_values_lst)
#https://stackoverflow.com/questions/11791568/what-is-the-most-pythonic-way-to-exclude-elements-of-a-list-that-start-with-a-sp
#gets unique values from a lst given
```

```python
def get_unique_column_values(lst_of_lst, index_column):
    # return funct(func2(lst_of_lst, index_column))
    return get_unique_values(get_column_values(lst_of_lst, index_column))
#gets unique values from a lst of values from a requested column
```

```python
def homeruns(lst_of_lst, index_column):
    max_percentile = []
    ninty_nine_percentile = []
    fifty_percentile = []
    year = get_unique_column_values(lst_of_lst, BattingData.year) # list of year with no one repeating
    for i in year:
        home_runs_in_year = get_matching_rows(lst_of_lst, BattingData.year, i) # list of values within the year
        home_runs = get_column_values(home_runs_in_year, BattingData.home_runs) # getting the number of homeruns
        home_runs_sorted = sorted(home_runs) #sorting the list of homeruns each year is a different list
        percentile_five = len(home_runs_sorted) * 0.50 #50th percentile index
        percentile_99 = len(home_runs_sorted) * .99 #99th percentile index
        max_percentile.append(home_runs_sorted[len(home_runs_sorted) - 1])
        ninty_nine_percentile.append(home_runs_sorted[int(percentile_99 + 1)]) # + 1 to round up and make it an interger intstead of a float
        fifty_percentile.append(home_runs_sorted[int(percentile_five + 1)])
    lst_of_labels = ["Maximum", "99th Percentile", "50th Percentile"]
    plt.bar(year, max_percentile)
    plt.bar(year, ninty_nine_percentile)
    plt.bar(year, fifty_percentile)
    plt.legend(lst_of_labels)
    plt.xlabel("Year")
    plt.ylabel("Number of Homeruns")
    plt.title("Homeruns per year by player")
    plt.show()
```

```python
def print_shape_of_data(file):
    # okay if you pass in the filename as a parameter to this function
    data = read_data(file)  
    
    rows = len(data)
    cols = len(data[0])

    # make sure all rows have the same number of columns
    same_number_columns = True
    for row in data:
        if len(row) != cols:
            same_number_columns = False

    if not same_number_columns:
        print("Some rows have a different number of columns!")

    # make sure each column is the correct type
    overall_type = True    
    str_cols = [BattingData.player_id, BattingData.first_name, BattingData.last_name, BattingData.team_name, BattingData.position]
    for column in range(0, cols):
        col_vals = get_column_values(data, column)
        correct_type = True
        for val in col_vals:
            if ((column in str_cols) and type(val) is not str) or (column not in str_cols and type(val) is not int):
                correct_type = False

        if not correct_type:
            print("Column " + str(column) + " contains values that are not the right type!")
            print(col_vals[:10]) # print the first 10 values
            overall_type = False


    if overall_type and same_number_columns:
        print("Rows: " + str(rows))
        print("Cols: " + str(cols))

        if rows == 74750 and cols == 20:
            print("Congrats! You have the correct shape of data!")
```
